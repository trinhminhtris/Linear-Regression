```python
import pandas as pd
url_1 = 'iris.csv'
data_1 = pd.read_csv(url_1)
data_1.head()
```

```python
import matplotlib.pyplot as plt
data_1.hist()
plt.show()
```

```python
correlation_1 = data_1.corr()
print(correlation_1)
```

```python
import seaborn as sns
sns.heatmap(correlation_1, annot = True)
```

```python
sns.pairplot(data_1)
```

```python
from sklearn import linear_model
model_1 = linear_model.LinearRegression()
model_1.fit(X = pd.DataFrame(data_1['petal.length']), y = data_1['petal.width'])
print('Hệ số intercept là: ', model_1.intercept_)
print('Hệ số coefficient là: ', model_1.coef_)
print('Độ chính xác của mo hình là: ', model_1.score(X = pd.DataFrame(data_1['petal.length']), y = data_1['petal.width']))
```

```python
y_pred_1 = model_1.predict(X = pd.DataFrame(data_1['petal.length']))
data_1.plot(kind = 'scatter', x = 'petal.length', y = 'petal.width', figsize = (9,9),
          color = 'black', xlim = (0,8))
plt.plot(data_1['petal.length'], y_pred_1, color = 'blue')
```

```python
# Model bậc 2
model_1 = linear_model.LinearRegression()
X = pd.DataFrame([data_1['petal.length'], data_1['petal.length']**2]).T
model_1.fit(X, y = data_1['petal.width'])
print('Intercept của mô hình: ', model_1.intercept_)
print('Coefficient của mô hình: ', model_1.coef_)
print('Độ chính xác của mô hình: ', model_1.score(X, y = data_1['petal.width']))
```

```python
import numpy as np
range_1 = np.arange(1, 7, 0.1)
X_range_1 = pd.DataFrame([range_1, range_1**2]).T
y_pred_1 = model_1.predict(X = X_range_1)
data_1.plot(kind = 'scatter', x = 'petal.length', y = 'petal.width', figsize = (9,9),
          color = 'black', xlim = (0,8))
plt.plot(range_1, y_pred_1, color = 'blue')
```

```python
# Model bậc 6
model_1 = linear_model.LinearRegression()
X = pd.DataFrame([data_1['petal.length'], data_1['petal.length']**2, data_1['petal.length']**3,
                  data_1['petal.length']**4, data_1['petal.length']**5, data_1['petal.length']**6]).T
model_1.fit(X, y = data_1['petal.width'])
print('Intercept của mô hình: ', model_1.intercept_)
print('Coefficient của mô hình: ', model_1.coef_)
print('Độ chính xác của mô hình: ', model_1.score(X, y = data_1['petal.width']))
```

```python
import numpy as np
range_1 = np.arange(1, 7, 0.1)
X_range_1 = pd.DataFrame([range_1, range_1**2, range_1**3, range_1**4, range_1**5, range_1**6]).T
y_pred_1 = model_1.predict(X = X_range_1)
data_1.plot(kind = 'scatter', x = 'petal.length', y = 'petal.width', figsize = (9,9),
          color = 'black', xlim = (0,8))
plt.plot(range_1, y_pred_1, color = 'blue')
```
